## Missing Data

Numeric columns with missing data (NaN value) are substituted with median value

### Substitute  missing values in msamd_name with value based on value in census_tract_number

```python
# Create a dataframe with columns census_tract_number and msamd_name by filtering on non-null values from column msamd_name
census_msamd_map = WA_HomeLoan_df[~WA_HomeLoan_df.msamd_name.isnull()][['census_tract_number', 'msamd_name']].drop_duplicates()
census_msamd_map.head()
```

```python
# convert census_tract_number and msamd_name map dataframe to dictionary
census_msamd_map_dict = dict(zip(census_msamd_map.census_tract_number, census_msamd_map.msamd_name))
```

```python
# Function to replace missing value in msamd_name with census_tract_number
def replace_msamdname(row):
    #for index, row in WA_HomeLoan_df.iterrows():
    if not pd.isnull(row.msamd_name):
        return row.msamd_name
    else:
        if row.census_tract_number in census_msamd_map_dict:
            row.msamd_name = census_msamd_map_dict[row.census_tract_number]
            return row.msamd_name
        else:
            row.msamd_name = 'Unknown'
            return row.msamd_name
```

```python
#  Substitute missing values in msamd_name with value based on value in census_tract_number
WA_HomeLoan_df.msamd_name = WA_HomeLoan_df.apply(replace_msamdname, axis=1)
```

## Outliers

```python
#functions for removing outliers
def removeOutliers(columnName):
    q25 = columnName.quantile(0.25)
    q75 = columnName.quantile(0.75)
    iqr =  q75 - q25
    cutoff = iqr*1.5
    lower = q25 - cutoff
    higher = q75 + cutoff
    columnName.mask(columnName < lower, other = lower, inplace=True)
    columnName.mask(columnName > higher, other = higher, inplace=True)
```

```python
# Remove outliers
removeOutliers(WA_HomeLoan_df.tract_to_msamd_income)
removeOutliers(WA_HomeLoan_df.population)
removeOutliers(WA_HomeLoan_df.minority_population)
removeOutliers(WA_HomeLoan_df.number_of_owner_occupied_units)
removeOutliers(WA_HomeLoan_df.number_of_1_to_4_family_units)
removeOutliers(WA_HomeLoan_df.loan_amount_000s)
removeOutliers(WA_HomeLoan_df.applicant_income_000s)
```
