```python
inPatient_df = pd.read_csv(inpatient_filename)
outpatient_df = pd.read_csv(outpatient_filename)
```

```python
def read_append_df(filepath, df_dtype):
    csv_files = glob.glob(filepath)
    data_list = []
    for filename in csv_files:
        df_data = pd.read_csv(filename, dtype=df_dtype, header=1)
        data_list.append(df_data)
    return pd.concat(data_list)
```

```python
partdprescriber_df = read_append_df(partdprescriber_filename, partdprescriber_dtype)
```
