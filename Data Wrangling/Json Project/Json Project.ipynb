```python
#import library
import pandas as pd
import json
from pandas.io.json import json_normalize
import numpy as np
```

```python
#Read Dataset into a frame and inspect first 5 rows
file_name = "E:/Data Science/SpringBoard DS/Data Wrangling Project/Json Project/data_wrangling_json/data_wrangling_json/data/world_bank_projects.json"
df = pd.read_json(file_name)
df.head()
```

```python
#inspect all the columns
pd.set_option('display.max_columns', 50)
df.head()
```

```python
#inspect the Dataframe any missing values
df.info()
```

## 1. Top 10 Countries with Most Projects 

```python
# Store top 10 countries with most project to 'top_10_countries'
top_10_countries = df.countryname.value_counts().head(10)
top_10_countries
```

## 2. Find the top 10 major Project Themes

```python
#Load dataset and inspect the data
with open(file_name, 'r') as json_file:
    json_data = json.load(json_file)

print(json_data[0])
```

```python
#data is nested, flatten the data using column mjtheme_namecode and inspect dataframe
df_json = json_normalize(json_data, 'mjtheme_namecode')
df_json.head()
```

```python
# Top 10 Themes with missing value
top_10_themes = df_json.name.value_counts().head(10)
top_10_themes
```

## 3.  Create a dataframe with the missing names filled in for column 'mjtheme_namecode'

```python
#Sort and replace with NaN
df_json_sorted=df_json.sort_values(by=['code', 'name'])
bool_data = df_json_sorted['name'] == ''
df_json_sorted['name'][bool_data] = np.nan
```

```python
#inspect first 15 rows to verify NaN values
df_json_sorted.head(15)
```

```python
#back fill the sorted Dataframe and store in into 'df_filledIn'
df_filledIn = df_json_sorted.fillna(method = 'bfill')
#inspect 'df_filledIn'
df_filledIn.head(15)
```

## Top 10 Themes after filling Missing values

```python
# Top 10 themes after filling the missed in values
df_filledIn.name.value_counts().head(10)
```
