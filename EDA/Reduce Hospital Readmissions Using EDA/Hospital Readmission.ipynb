# Hospital Readmissions Data Analysis and Recommendations for Reduction

### Background
In October 2012, the US government's Center for Medicare and Medicaid Services (CMS) began reducing Medicare payments for Inpatient Prospective Payment System hospitals with excess readmissions. Excess readmissions are measured by a ratio, by dividing a hospital’s number of “predicted” 30-day readmissions for heart attack, heart failure, and pneumonia by the number that would be “expected,” based on an average hospital with similar patients. A ratio greater than 1 indicates excess readmissions.

```python
# deal with missing and inconvenient portions of data 
clean_hospital_read_df = hospital_read_df[hospital_read_df['Number of Discharges'] != 'Not Available']
clean_hospital_read_df.loc[:, 'Number of Discharges'] = clean_hospital_read_df['Number of Discharges'].astype(int)
clean_hospital_read_df = clean_hospital_read_df.sort_values('Number of Discharges')
```

# B. Provide support for your arguments and your own recommendations with a statistically sound analysis

```python
hospital_df = clean_hospital_read_df[['Number of Discharges','Excess Readmission Ratio']].dropna()
```

```python
hospital_df.columns = ['discharges','readmissionratio']
```

```python
hospital_df.discharges = hospital_df.discharges.astype(int)
```

## 1. Setup an appropriate hypothesis test.

Null Hypothesis test Ho : There is no correlation between number of discharges and excess readmission ratio.

Alternate Hyposthesis Ha : There is a correlation between number of discharges and excess readmission ratio.

## 2. Compute and report the observed significance value (or p-value).

```python
corr = hospital_df.discharges.corr(hospital_df.readmissionratio)
corr
```

```python
perm_reps = np.empty(10000)
for i in range(10000):
    perms = np.random.permutation(hospital_df.discharges)
    perm_reps[i] = np.corrcoef(perms,hospital_df.readmissionratio)[0,1]
p = np.sum(perm_reps <= corr)/10000
print('p value:',p)
```

** *P value is zero so we can reject null hypothesis. * **

```python
sns.boxplot(perm_reps)
plt.show()
```

```python
sns.distplot(perm_reps)
#plt.hist(perm_reps, bins=100)
plt.show()
```

```python
conf_int = np.percentile(perm_reps,[0.5,99.5])
print("95% confidence interval is : {}".format(conf_int))
```

Null hypothesis can be rejected since P-value is zero and population correlation falls outside the 95% confidence interval.

## 3. Report statistical significance for  αα  = .01.


```python
conf_int_99 = np.percentile(perm_reps,[0.01,99.9])
print("99% confidence interval is : {}".format(conf_int_99))
```
